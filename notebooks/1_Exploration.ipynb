```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load dữ liệu đã làm sạch
df = pd.read_csv('../data/processed/foody_clean_master.csv')
print(f"Tổng số review: {len(df)}")
df.head()
```

### 1. Phân bố nhãn cảm xúc (Sentiment Distribution)

```python
plt.figure(figsize=(8, 6))
sentiment_counts = df['sentiment_label'].value_counts()
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', 
        colors=['#4A90D9', '#F4A460', '#E74C3C'], startangle=140)
plt.title('Tỷ lệ các nhãn cảm xúc trong tập dữ liệu')
plt.axis('equal')
plt.show()
```

### 2. Phân bố Rating
Xem xét sự phân bổ điểm số từ 1 đến 10.

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='rating', palette='viridis')
plt.title('Phân bố điểm đánh giá (Rating)')
plt.xlabel('Điểm số')
plt.ylabel('Số lượng')
plt.show()
```

### 3. Thống kê theo Vùng miền (Region)

```python
region_stats = df.groupby('region')['rating'].mean().sort_values(ascending=False)
print("Điểm đánh giá trung bình theo vùng miền:")
print(region_stats)
```
